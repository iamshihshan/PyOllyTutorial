# [Matplotlib](https://matplotlib.org/)

* Guide of how to do data visualization with Matplotlib module

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assign your data as list like below
year = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
income = [102,385,265,423,465,432,462,435,521,572]
```

```python
# Yearly income Plot in red
plt.plot(year, income, color='red')
plt.title('My First Matplot')

plt.show()
```

## plt.axes()

* Show your graph beside each other
* plt.axes([x0_location, y0_location ,  Width, Height])

```python
plt.axes([0, 0, 0.50, 1])
plt.plot(year, income, color='red')

plt.axes([0.60, 0, 0.50, 1])
plt.plot(year, income, 'bo')

plt.axes([1.20, 0, 0.50, 1])
plt.plot(year, income, 'r+')

plt.show()
```

## plt.subplot()
* Show your graph beside each other
* plt.subplot(r, c, loc)
    * r : row
    * c : column
    * loc : location  from left top --> right

|(2,3,1)|(2,3,2)|(2,3,3)|
|-------|-------|-------|
|(2,3,4)|(2,3,5)|(2,3,6)|

```python
for i, c in [(1,'blue'), (2, 'red'), (3, 'Orange'), (4, 'purple'), (5, 'Gray'), (6,'Green')]:
    # Create a figure with 2x3 subplot
    plt.subplot(2, 3, i) 
    # Plot
    plt.plot(year, income, color=c)
    # Add title
    plt.title('My Yearly Income')


# improve the spacing between subplots
plt.tight_layout()
plt.show()
```

## plt.xlim(), plt.ylim() vs plt.axis()
* Zoom in the specific part of your graph
* plt.xlim((lower_num, upper_num))
* plt.ylim((lower_num, upper_num))
* plt.axis((x_lower_num, x_upper_num, y_lower_num, y_upper_num))

```python
# For xlim(), ylim()
plt.subplot(2,1,1)
plt.plot(year,income, color='blue')
# Add the axis labels
plt.xlabel('Year')
plt.ylabel('Income (k)')
# Add title
plt.title('My Income \n Zoom In with xlim(), ylim()')
# Set the x-axis range
plt.xlim((2010,2017))
# Set the y-axis range
plt.ylim((400,600))


# For axis()
plt.subplot(2,1,2)
plt.plot(year,income, color='red')
# Add the axis labels
plt.xlabel('Year')
plt.ylabel('Income (k)')
# Add title
plt.title('My Income  \n Zoom In with axis()')
# Set the x-axis & y-axis both
plt.axis((2006, 2017, 50, 650))


plt.tight_layout()
plt.show()
```

## plt.legend()
* legend(col='XXX')
    * XXX : lower center/upper right/center/right/best/center/right...

```python
# Assign another data as list like below
payment = [102,200,535,420,250,350,324,450,500,372]
```

```python
# Label your data source
plt.plot(year, income, color='blue', label='Income')
plt.plot(year, payment, color='red', label='Payment')
plt.title('My Income & Payment \n legend(loc=\'XXX\')')

# Give legnet for your graph
plt.legend(loc='lower right')

plt.show()
```

## plt.annotate()
* Give annotate for your graph
* plt.annotate(s='XXXX', xy=(x_loc, y_loc), xytext=(x_text_loc+1, y_text_loc+1), arrowprops=dict(facecolor='black'))

```python
# Label your data source
plt.plot(year, income, color='blue', label='Income')
plt.plot(year, payment, color='red', label='Payment')
plt.title('My Income & Payment \n annotate()')

# Give legnet for your graph
plt.legend(loc='lower right')

# Compute the maximum income location : income_max, y_max
income_max = max(income)
yr_max = year[np.argmax(income)]


# Add a black arrow annotation
plt.annotate(s='Max', xy=(yr_max, income_max), xytext=(yr_max+1, income_max+1), arrowprops=dict(facecolor='black'))
plt.show()
```

## Histogram Plot

```python
x = np.random.randn(1000)
plt.hist(x, bins=10, alpha=0.5)
plt.title('Histogram Plot', size=20)
```

```python
x1 = np.random.randn(1000)
x2 = np.sin(x1)
x3 = np.cos(x1)

plt.hist(x1, bins=10, alpha=0.5, label=['x'])
plt.hist(x2, bins=5, alpha=0.5, label=['sin(x)'])
plt.hist(x3, bins=5, alpha=0.5, label=['cos(x)'])
plt.legend()
plt.title('Histogram Plot', size=20)
```

## Scatter Plot

```python
x = np.random.randn(100)
y = np.sin(x)
plt.plot(x, y, 'go', alpha=0.5)
plt.title('Scatter Plot', size=20)
```

```python
x = np.random.randn(100)
y = np.sin(x)
plt.plot(x, y, 'ro', alpha=0.5, label='sin(x)')
plt.plot(x, -y, 'm*', alpha=0.5, label='-sin(x)')
plt.plot([-3, 3],[0, 0], 'k-')
plt.legend()
plt.title('Scatter Plot', size=20)
```

## Bar Plot

```python
x = ['Python', 'SQL', 'R', 'SAS', 'Scala']
x_loc = np.arange(len(x))
y = [10, 8, 6, 3,6]
plt.bar(x_loc, y, alpha=0.5)
plt.xticks(x_loc, x)
plt.title('Bar Plot', size=20)
```

```python
x = ['Python', 'SQL', 'R', 'SAS', 'Scala']
x_loc = np.arange(len(x))
y1 = np.array([10, 8, 6, 3,6])
y2 = np.array([2, 5, 5, 1,2])

plt.bar(x_loc, y1, alpha=0.5, label='y1')
plt.bar(x_loc, y2, alpha=0.5, bottom=y1, label='y2')
plt.xticks(x_loc, x)
plt.legend()

plt.title('Bar Plot', size=20)
```

## Box Plot

```python
x1 = np.random.randn(100)
x2 = np.sin(x1)
x3 = np.cos(x1)
plt.boxplot([x1,x2,x3], labels=['A','B','C'])
```

## Pie Chart

```python
sizes = [360, 130, 245, 210, 160]
 
plt.pie(sizes,
        explode=(0.1, 0, 0, 0, 0),
        labels=['Python', 'SQL', 'R', 'SAS', 'Scala'],
        autopct='%1.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.legend()
plt.show()
```
